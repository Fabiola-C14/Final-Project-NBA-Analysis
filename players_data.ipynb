```python
# Dependencies
import requests
import json
import pandas as pd
from config import api_key
from pprint import pprint
```

```python
url="https://api.sportsdata.io/v3/nba/scores/json/Players?"
```

```python
query_url = f"{url}key={api_key}"
```

```python
response = requests.get(query_url)
response_json = response.json()
pprint(json.dumps(response_json, indent=4, sort_keys=True))
```

```python
response_json[0]
```

```python
Team=[]
Jersey=[]
PositionCategory=[]
Position=[]
FirstName=[]
LastName=[]
Height=[]
Weight=[]
BirthDate=[]
BirthCity=[]
BirthState=[]
BirthCountry=[]
College=[]
Salary=[]
PhotoUrl=[]
Experience=[]
PlayerID=[]

for entry in response_json:

    Team.append(entry['Team'])
    Jersey.append(entry['Jersey'])
    PositionCategory.append(entry['PositionCategory'])
    Position.append(entry['Position'])
    FirstName.append(entry['FirstName'])
    LastName.append(entry['LastName'])
    Height.append(entry['Height'])
    Weight.append(entry['Weight'])
    BirthDate.append(entry['BirthDate'])
    BirthCity.append(entry['BirthCity'])
    BirthState.append(entry['BirthState'])
    BirthCountry.append(entry['BirthCountry'])
    College.append(entry['College'])
    Salary.append(entry['Salary'])
    PhotoUrl.append(entry['PhotoUrl'])
    Experience.append(entry['Experience'])
    PlayerID.append(entry['PlayerID'])
    
    
```

```python
player_data=pd.DataFrame({'Player ID':PlayerID, 'First Name':FirstName, 'Last Name':LastName, 'Team':Team, 'Jersey':Jersey, 'Position Category':PositionCategory, 'Position':Position, 'Height': Height, 'Weight': Weight, 'Birth Date':BirthDate, 'Birth City':BirthCity, 'Birth State':BirthState, 'Birth Country':BirthCountry,'College':College, 'Salary':Salary, 'Photo Url':PhotoUrl, 'Experience': Experience, 'Player ID':PlayerID})

player_data
```

```python
player_data.to_csv("players.csv",index=False)
```
